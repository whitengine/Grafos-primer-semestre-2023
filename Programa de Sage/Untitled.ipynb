```python
#Ploteamos un grafo en particular usando la funcion Graph(diccionario) que lo que hace es agarrar un diccionario
#{1: [1,3],2:[4],3:[1,2,4],4: []}  donde cada key representa el vertice=key y tiene un valor igual a una lista con
#los vertices adyacentes.

plotG = []


@interact
def _(auto_update = True,button = True, selector = ["acyclic", "circular" , "ranked", "graphviz", "planar", "spring", "tree"]):
    global plotG
    
    G = graphs.CycleGraph(4)
    
    G.add_edges([(0,4) , (4,5), (5,2), (0 , 6) , (6,7) , (7,8) , (8,2)])
        
        
    plotG = G.plot(layout = selector)
    
    plotG.show()
```

```python
G = graphs.PathGraph(4)


for u,v,l in G.edges():
    if u == 2 and v == 3:
        G.set_edge_label(u,v,'$\\cdots$')
    else:
        G.set_edge_label(u,v,'')
        
G.relabel(lambda i: f'$x_{i}$' if i != 3 else '$x_k$')
        

plotG = G.plot(edge_labels = True)
plotG.show()
```
