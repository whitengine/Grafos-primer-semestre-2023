```python
#Creemos una funciÃ³n que agarra un grafo, lo dibuja y guarda el archivo en la carpeta de la "Clase n" y lo salva con el nombre
#Dibujo r.pdf donde r el el menor numero entero que todavia no se uso para indexar los dibujos de esa Clase.

#Sea G el grafo que creamos mÃ¡s arriba, entonces plot = G.plot().

import os

guardar = False           #esta variable decide si se puede guardar el archivo, empieza apagado para que cuando corramos el programa
#no descargue automaticamente

@interact
def Dibujar(clase_num = selector(['Clase ' + str(i) for i in range(1,21)],default = clase_default), auto_update=False, ):
    global guardar
    
    if guardar: #si guardar es falzo cuando compilamos entonces no descarga.
        path = os.path.join("Figuras", clase_num)
        
        #este bloque encuentra cual es el primer nombre "Clase x" disponible
        x = 1
        Dibujo_path = os.path.join("..",path, "Dibujo " + str(x) + ".pdf")
        while os.path.exists(Dibujo_path):   
            x = x + 1
            Dibujo_path = os.path.join("..",path, "Dibujo " + str(x) + ".pdf")
        
        #plot = G.plot() de mas arriba del jupyter
        plot.save(Dibujo_path) #guarda el archivo.
        print(Dibujo_path)
        
    guardar = True #recien descarga la priemra vez que actualizamos jupyter. i.e. cuando tocamos el boton "Run Interact"

    
```

```python
#Hagamos una lista de funciones nuevas para grafos en sage

def intersecar_grafos(G,H):
    V = list(Set(G.vertices()).intersection(Set(H.vertices())))
    #estos son los vertices de la interseccion
    
    
    #Luego hacemos un diccionario con las aristas y las conectamos
    #a verties adyacentes siempre y cuando la arista pertenezca 
    #a la interseccion de las aristas
    
    Dict = {i : [j for j in V if (i,j,None) in G.edges() and (i,j,None) in H.edges()] for i in V}
    
    return Graph(Dict)

```
